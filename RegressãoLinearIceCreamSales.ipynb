```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("athiramolps/ice-cream-sales-and-temperature")

print("Path to dataset files:", path)
```

```python
import os

# List the contents of the dataset directory to find the correct filename
print(os.listdir(path))
```

```python
import pandas as pd
```

```python
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Construct the full path to the CSV file
dataset_file_path = os.path.join(path, 'Ice Cream Sales and Temperature.csv')

# Load the dataset
df = pd.read_csv(dataset_file_path)

# Display the first 5 rows of the DataFrame
display(df.head())
```

```python
# Create a linear regression plot
plt.figure(figsize=(10, 6))
sns.regplot(x='Temperature (Celsius)', y='Ice Cream Sales', data=df)
plt.title('Linear Regression of Ice Cream Sales vs. Temperature')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Ice Cream Sales (USD)')
plt.grid(True)
plt.show()
```
